```python
class node:
    def __init__(self):
        self.previous = None
        self.data = None
        self.next = None
        
def insert(head, data):
    n = node()
    q = head
    if q == None:
        head = n
        n.previous = None
        n.data = data
        n.next = None
    else:
        while q.next != None:
            q = q.next
        q.next = n
        n.previous = q
        n.data = data
        n.next = None
    return head

def display(head):
    q = head
    if q == None:
        print("List is empty")
    else:
        while q != None:
            print(q.data, end='\t')
            q = q.next
            
def displayReverse(head):
    q = head
    if q == None:
        print("List is empty")
    else:
        while q.next != None:
            q = q.next
        while q != None:
            print(q.data, end='\t')
            q = q.previous
            
def delete(head, data):
    q = head
    if q == None:
        print("List is empty")
    else:
        if q.data == data:
            head = q.next
            del q
        else:
            while q.next.data != data:
                q = q.next
            q.next, temp = q.next.next, q.next
            if q.next != None:
                q.next.previous = q
            del temp
        return head
    
head = None
head = insert(head, 5)
head = insert(head, 2)
head = insert(head, 7)
head = insert(head, 3)
head = insert(head, 9)
head = insert(head, 1)

print("Ascending order: ", end='')
display(head)

print()
print("Descending order: ", end='')
displayReverse(head)

head = delete(head, 7)
print()
print("Deleted 7: ", end='')
display(head)

head = delete(head, 5)
print()
print("Deleted 5: ", end='')
display(head)

head = delete(head, 1)
print()
print("Deleted 1: ", end='')
display(head)
```
